```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
import re
import PyPDF2

# === Filenames ===
raw_file = "calcard_2025-11-07_16_28_40.raw"
pdf_file = "Mira220_DS000642_9-00.pdf"

# === Step 1: Extract width and height from the datasheet ===
def extract_resolution_from_pdf(pdf_path):
    with open(pdf_path, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        text = ""
        for page in reader.pages:
            text += page.extract_text()

    # Look for resolution patterns like 1920x1080, 1600 x 1300, etc.
    match = re.search(r"(\d{3,5})\s*[x]\s*(\d{3,5})", text)
    if match:
        width, height = map(int, match.groups())
        print(f"Detected resolution: {width}x{height}")
        return width, height
    else:
        raise ValueError("Could not detect resolution from datasheet text.")

width, height = extract_resolution_from_pdf(pdf_file)

# === Step 2: Define 4x4 pixel pattern ===
pattern = np.array([
    ['G', 'R', 'G', 'B'],
    ['I', 'G', 'I', 'G'],
    ['G', 'B', 'G', 'R'],
    ['I', 'G', 'I', 'G']
])

# === Step 3: Load RAW data ===
raw_path = Path(raw_file)
data = np.fromfile(raw_path, dtype=np.uint16)
expected = width * height
if data.size != expected:
    raise ValueError(f"Unexpected file size. Got {data.size} pixels, expected {expected}. Check sensor bit depth or resolution.")

raw = data.reshape((height, width))

# === Step 4: Create masks ===
mask_R = np.zeros_like(raw, dtype=bool)
mask_G = np.zeros_like(raw, dtype=bool)
mask_B = np.zeros_like(raw, dtype=bool)
mask_I = np.zeros_like(raw, dtype=bool)

for y in range(4):
    for x in range(4):
        ch = pattern[y, x]
        if ch == 'R':
            mask_R[y::4, x::4] = True
        elif ch == 'G':
            mask_G[y::4, x::4] = True
        elif ch == 'B':
            mask_B[y::4, x::4] = True
        elif ch == 'I':
            mask_I[y::4, x::4] = True

# === Step 5: Normalize and interpolate missing pixels ===
def normalize_masked(raw, mask):
    img = np.zeros_like(raw, dtype=np.float32)
    img[mask] = raw[mask].astype(np.float32)
    img = cv2.normalize(img, None, 0, 1, cv2.NORM_MINMAX)
    return img

R = normalize_masked(raw, mask_R)
G = normalize_masked(raw, mask_G)
B = normalize_masked(raw, mask_B)

R_demosaic = cv2.inpaint(R, (~mask_R).astype(np.uint8), 3, cv2.INPAINT_NS)
G_demosaic = cv2.inpaint(G, (~mask_G).astype(np.uint8), 3, cv2.INPAINT_NS)
B_demosaic = cv2.inpaint(B, (~mask_B).astype(np.uint8), 3, cv2.INPAINT_NS)

rgb_image = np.clip(np.stack([R_demosaic, G_demosaic, B_demosaic], axis=2), 0, 1)

# === Step 6: RGB-IR mosaic visualization ===
mosaic_rgbir = np.zeros((height, width, 3), dtype=np.float32)
mosaic_rgbir[..., 0] = (raw * (mask_R | mask_I)).astype(np.float32)
mosaic_rgbir[..., 1] = (raw * mask_G).astype(np.float32)
mosaic_rgbir[..., 2] = (raw * (mask_B | mask_I)).astype(np.float32)
mosaic_rgbir = cv2.normalize(mosaic_rgbir, None, 0, 1, cv2.NORM_MINMAX)

# === Step 7: Save images ===
rgb_path = raw_path.with_name("calcard_RGB.png")
mosaic_path = raw_path.with_name("calcard_RGBIR_mosaic.png")

plt.imsave(rgb_path, rgb_image)
plt.imsave(mosaic_path, mosaic_rgbir)

# === Step 8: Plot ===
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("RGB Demosaiced Image")
plt.imshow(rgb_image)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("RGB-IR Mosaic")
plt.imshow(mosaic_rgbir)
plt.axis("off")

plt.show()

print(f"Saved:\n - {rgb_path}\n - {mosaic_path}")
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path

# --- Configuration ---
width = 1600
height = 1400
bit_depth = 12
raw_filename = "calcard_2025-11-07_16_28_40.raw"

# --- Load RAW file ---
path = Path.cwd() / raw_filename
raw = np.fromfile(path, dtype=np.uint16) & 0x0FFF
if raw.size != width * height:
    raise ValueError(f"Unexpected RAW size: {raw.size}, expected {width*height}")
raw = raw.reshape((height, width))

# --- Mira220 RGB-IR 44 CFA pattern (approx from Fig. 35) ---
# pattern layout repeating every 4x4 block:
# R  Gr  R  IR
# Gr B  Gr  IR
# R  Gr  R  IR
# Gr B  Gr  IR

R  = raw[0::4, 0::4]        # main red
R2 = raw[0::4, 2::4]        # secondary red
G1 = raw[0::4, 1::4]
G2 = raw[1::4, 0::4]
G3 = raw[1::4, 2::4]
G4 = raw[3::4, 1::4]
B1 = raw[1::4, 1::4]
B2 = raw[3::4, 1::4]
IR1 = raw[0::4, 3::4]
IR2 = raw[1::4, 3::4]
IR3 = raw[2::4, 3::4]
IR4 = raw[3::4, 3::4]

# Average same-color groups to reduce aliasing
R_full  = cv2.resize(((R + R2) / 2).astype(np.float32), (width, height))
G_full  = cv2.resize(((G1 + G2 + G3 + G4) / 4).astype(np.float32), (width, height))
B_full  = cv2.resize(((B1 + B2) / 2).astype(np.float32), (width, height))
IR_full = cv2.resize(((IR1 + IR2 + IR3 + IR4) / 4).astype(np.float32), (width, height))

# Normalize to 8-bit for viewing
def norm8(img):
    return np.clip((img / (2**bit_depth - 1)) * 255, 0, 255).astype(np.uint8)

rgb = cv2.merge([norm8(R_full), norm8(G_full), norm8(B_full)])
ir  = norm8(IR_full)

# --- Create pseudo RGB-IR mosaic (RGB + IR overlay) ---
rgbir_mosaic = rgb.copy()
rgbir_mosaic[..., 0] = cv2.addWeighted(rgb[..., 0], 0.7, ir, 0.3, 0)
rgbir_mosaic[..., 2] = cv2.addWeighted(rgb[..., 2], 0.7, ir, 0.3, 0)

# --- Save results ---
cv2.imwrite(str(path.parent / "mira220_rgb.png"), cv2.cvtColor(rgb, cv2.COLOR_RGB2BGR))
cv2.imwrite(str(path.parent / "mira220_rgbir.png"), cv2.cvtColor(rgbir_mosaic, cv2.COLOR_RGB2BGR))

# --- Optional plot ---
plt.figure("RGB image"); plt.imshow(rgb); plt.axis("off")
plt.figure("RGB-IR mosaic"); plt.imshow(rgbir_mosaic); plt.axis("off")
plt.show()
```

```python
import cv2
import numpy as np
from pathlib import Path
from matplotlib import pyplot as plt

# --- Configuration ---
ref_path = Path("IMG_0142.jpg")        # iPhone reference
cam_path = Path("mira220_rgb.png")      # Mira220 image
output_path = Path("mira220_rgb_corrected.png")

# --- Load images ---
ref = cv2.imread(str(ref_path), cv2.IMREAD_UNCHANGED)
cam = cv2.imread(str(cam_path), cv2.IMREAD_UNCHANGED)
if ref is None or cam is None:
    raise FileNotFoundError("Could not open reference or camera image.")
ref = cv2.cvtColor(ref, cv2.COLOR_BGR2RGB)
cam = cv2.cvtColor(cam, cv2.COLOR_BGR2RGB)

# --- Detect color checker chart automatically ---
def detect_color_chart(image):
    # Convert to grayscale and threshold
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    gray = cv2.GaussianBlur(gray, (5, 5), 0)
    edges = cv2.Canny(gray, 50, 150)
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    chart_contour = None
    max_area = 0
    for cnt in contours:
        peri = cv2.arcLength(cnt, True)
        approx = cv2.approxPolyDP(cnt, 0.02 * peri, True)
        area = cv2.contourArea(approx)
        if len(approx) == 4 and area > max_area:
            chart_contour = approx
            max_area = area
    if chart_contour is None:
        raise RuntimeError("Chart not detected automatically.")
    pts = np.array([p[0] for p in chart_contour], dtype=np.float32)
    return pts

def warp_chart(image, pts, w=600, h=400):
    # Sort points and warp to a flat top-down view
    rect = np.zeros((4, 2), dtype="float32")
    s = pts.sum(axis=1)
    rect[0] = pts[np.argmin(s)]
    rect[2] = pts[np.argmax(s)]
    diff = np.diff(pts, axis=1)
    rect[1] = pts[np.argmin(diff)]
    rect[3] = pts[np.argmax(diff)]
    dst = np.array([[0, 0], [w - 1, 0], [w - 1, h - 1], [0, h - 1]], dtype="float32")
    M = cv2.getPerspectiveTransform(rect, dst)
    return cv2.warpPerspective(image, M, (w, h))

# --- Warp both charts ---
ref_pts = detect_color_chart(ref)
cam_pts = detect_color_chart(cam)
ref_warp = warp_chart(ref, ref_pts)
cam_warp = warp_chart(cam, cam_pts)

# --- Define 6x4 grid (24 patches) like X-Rite color checker ---
n_rows, n_cols = 4, 6
patch_w = ref_warp.shape[1] // n_cols
patch_h = ref_warp.shape[0] // n_rows

def extract_patches(img):
    patches = []
    for r in range(n_rows):
        for c in range(n_cols):
            x0 = c * patch_w + patch_w // 4
            y0 = r * patch_h + patch_h // 4
            x1 = x0 + patch_w // 2
            y1 = y0 + patch_h // 2
            patch = img[y0:y1, x0:x1]
            patches.append(np.mean(patch.reshape(-1, 3), axis=0))
    return np.array(patches)

ref_colors = extract_patches(ref_warp)
cam_colors = extract_patches(cam_warp)

# --- Compute 3x3 color correction matrix via least squares ---
A = cam_colors
B = ref_colors
A = np.hstack([A, np.ones((A.shape[0], 1))])  # add bias column
X, _, _, _ = np.linalg.lstsq(A, B, rcond=None)
M = X[:3, :].T
bias = X[3, :]

print("Color correction matrix:\n", M)
print("Bias:", bias)

# --- Apply correction ---
cam_float = cam.astype(np.float32)
corrected = np.dot(cam_float, M.T) + bias
corrected = np.clip(corrected, 0, 255).astype(np.uint8)

# --- Save and show results ---
cv2.imwrite(str(output_path), cv2.cvtColor(corrected, cv2.COLOR_RGB2BGR))

plt.figure("Reference"); plt.imshow(ref); plt.axis("off")
plt.figure("Original Mira220"); plt.imshow(cam); plt.axis("off")
plt.figure("Corrected Mira220"); plt.imshow(corrected); plt.axis("off")
plt.show()
```
