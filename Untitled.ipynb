```python
"""
leaf_mira220_rgbir.py

Processing for Mira220 RGB-IR RAW frame captured with Arducam.

Assumptions:
- Raw file: 16-bit little-endian (RAW12 stored in 16 bits is typical).
- Resolution: 1600 x 1400 (Mira220 full frame).
- CFA pattern (4x4, repeating):

    Row 0: B  G  R  G
    Row 1: G  I  G  I
    Row 2: R  G  B  G
    Row 3: G  I  G  I

- "I" = IR.

You get:
- Horizontally flipped raw.
- Naive RGB demosaic (simple neighborhood interpolation).
- RGB-IR mosaic visualization.
"""

from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt

```

```python
# Raw file name
RAW_FILENAME = "calcard_2025-11-07_16_28_40.raw"
```

```python
# ----------------------------------------------------------------------
# Configuration
# ----------------------------------------------------------------------

# Mira220 full resolution
WIDTH = 1600
HEIGHT = 1400

# ----------------------------------------------------------------------
# Core utilities
# ----------------------------------------------------------------------

def load_raw_frame(
    filename: str = RAW_FILENAME,
    width: int = WIDTH,
    height: int = HEIGHT,
    dtype=np.uint16,
) -> np.ndarray:
    """
    Load a single RAW frame as a 2D numpy array [H, W].

    Assumes uncompressed, row-major, little-endian data.
    """
    path = Path(filename)
    if not path.exists():
        raise FileNotFoundError(f"Cannot find {filename} in current directory.")

    raw_1d = np.fromfile(path, dtype=dtype)

    expected = width * height
    if raw_1d.size != expected:
        raise ValueError(
            f"Size mismatch for {filename}: "
            f"expected {expected} pixels ({width}x{height}), "
            f"got {raw_1d.size}. "
            f"Adjust WIDTH/HEIGHT or dtype in this script to match your data."
        )

    raw_2d = raw_1d.reshape(height, width)
    return raw_2d


def flip_horizontal(frame: np.ndarray) -> np.ndarray:
    """
    Flip image horizontally.
    """
    return np.fliplr(frame)


# ----------------------------------------------------------------------
# Mira220 RGB-IR pattern handling
# Pattern (4x4, repeating):
#
#   [ B  G  R  G ]
#   [ G  I  G  I ]
#   [ R  G  B  G ]
#   [ G  I  G  I ]
#
# Indices are (row, col).
# ----------------------------------------------------------------------

def build_rgbir_masks(height: int, width: int):
    """
    Build boolean masks for R, G, B, IR based on the 4x4 Mira220 RGB-IR pattern.
    """
    y, x = np.indices((height, width))
    ym = y % 4
    xm = x % 4

    # Blue positions
    b_mask = ((ym == 0) & (xm == 0)) | ((ym == 2) & (xm == 2))

    # Red positions
    r_mask = ((ym == 0) & (xm == 2)) | ((ym == 2) & (xm == 0))

    # IR positions
    i_mask = ((ym == 1) & (xm == 1)) | ((ym == 1) & (xm == 3)) | \
             ((ym == 3) & (xm == 1)) | ((ym == 3) & (xm == 3))

    # All other non-IR, non-R, non-B in this pattern are G
    g_mask = ~(b_mask | r_mask | i_mask)

    return r_mask, g_mask, b_mask, i_mask


def split_rgbir_channels(raw: np.ndarray):
    """
    Split a single RAW frame into four sparse planes: R, G, B, IR.

    Each channel has the same shape as raw, with zeros where that channel
    is not sampled.
    """
    h, w = raw.shape
    r_mask, g_mask, b_mask, i_mask = build_rgbir_masks(h, w)

    r = np.zeros_like(raw, dtype=np.float32)
    g = np.zeros_like(raw, dtype=np.float32)
    b = np.zeros_like(raw, dtype=np.float32)
    ir = np.zeros_like(raw, dtype=np.float32)

    r[r_mask] = raw[r_mask]
    g[g_mask] = raw[g_mask]
    b[b_mask] = raw[b_mask]
    ir[i_mask] = raw[i_mask]

    return r, g, b, ir


def interpolate_channel_sparse(channel: np.ndarray) -> np.ndarray:
    """
    Simple interpolation for a sparse color plane.

    For pixels with value == 0, replace with the mean of non-zero neighbors
    in a 3x3 window. Pure NumPy, no SciPy or OpenCV required.
    """
    h, w = channel.shape
    vals = channel.astype(np.float32)
    mask_nonzero = (vals > 0).astype(np.float32)

    sum_neighbors = np.zeros_like(vals, dtype=np.float32)
    count_neighbors = np.zeros_like(vals, dtype=np.float32)

    # Collect contributions from 8-connected neighbors
    for dy in (-1, 0, 1):
        for dx in (-1, 0, 1):
            if dx == 0 and dy == 0:
                continue

            # destination indices
            y_dst = slice(max(0, dy), h + min(0, dy))
            x_dst = slice(max(0, dx), w + min(0, dx))

            # source indices
            y_src = slice(max(0, -dy), h - max(0, dy))
            x_src = slice(max(0, -dx), w - max(0, dx))

            shifted_vals = np.zeros_like(vals, dtype=np.float32)
            shifted_mask = np.zeros_like(mask_nonzero, dtype=np.float32)

            shifted_vals[y_dst, x_dst] = vals[y_src, x_src]
            shifted_mask[y_dst, x_dst] = mask_nonzero[y_src, x_src]

            sum_neighbors += shifted_vals
            count_neighbors += shifted_mask

    interp = vals.copy()
    missing = (vals == 0) & (count_neighbors > 0)
    interp[missing] = sum_neighbors[missing] / count_neighbors[missing]

    # If there are pixels with no neighbors, they stay 0. That should only occur
    # possibly on extreme borders.
    return interp


def demosaic_rgb_from_rgbir(raw: np.ndarray) -> np.ndarray:
    """
    Very simple demosaic for the Mira220 RGB-IR pattern.

    - Extract sparse R, G, B planes.
    - Interpolate each plane.
    - Stack into RGB.

    IR is ignored for the RGB reconstruction.
    """
    r_sparse, g_sparse, b_sparse, _ = split_rgbir_channels(raw)

    r_full = interpolate_channel_sparse(r_sparse)
    g_full = interpolate_channel_sparse(g_sparse)
    b_full = interpolate_channel_sparse(b_sparse)

    # Stack into [H, W, 3]
    rgb = np.stack([r_full, g_full, b_full], axis=-1)

    return rgb


def normalize_for_display(img: np.ndarray, clip_percentile: float = 99.0) -> np.ndarray:
    """
    Normalize image to 0..1 for display.

    Works for 2D or 3D images. Clips at a given percentile to reduce the
    influence of hot pixels / outliers.
    """
    img = img.astype(np.float32)

    if img.ndim == 3:
        flat = img.reshape(-1, img.shape[-1])
        hi = np.percentile(flat, clip_percentile, axis=0)
        lo = np.percentile(flat, 0.5, axis=0)
        # avoid division by zero
        scale = np.where(hi > lo, hi - lo, 1.0)
        out = (img - lo) / scale
    else:
        hi = np.percentile(img, clip_percentile)
        lo = np.percentile(img, 0.5)
        if hi == lo:
            hi = lo + 1.0
        out = (img - lo) / (hi - lo)

    return np.clip(out, 0.0, 1.0)


def build_rgbir_mosaic_vis(raw: np.ndarray) -> np.ndarray:
    """
    Build an RGB image visualizing the RGB-IR mosaic:

    - B sites: pure blue
    - R sites: pure red
    - G sites: pure green
    - IR sites: magenta (R+ B)

    Intensities are taken from the raw values per-site. No spectral
    calibration is applied.
    """
    r_sparse, g_sparse, b_sparse, ir_sparse = split_rgbir_channels(raw)

    # Normalize each sparse channel separately for visualization
    r_n = normalize_for_display(r_sparse)
    g_n = normalize_for_display(g_sparse)
    b_n = normalize_for_display(b_sparse)
    ir_n = normalize_for_display(ir_sparse)

    # Empty RGB canvas
    h, w = raw.shape
    vis = np.zeros((h, w, 3), dtype=np.float32)

    # Where R, G, B, IR exist, write them into RGB channels
    # Sparse maps are already zero elsewhere.
    vis[..., 0] += r_n  # R channel
    vis[..., 1] += g_n  # G channel
    vis[..., 2] += b_n  # B channel

    # Map IR to magenta: add to R and B channels
    vis[..., 0] += ir_n * 0.8
    vis[..., 2] += ir_n * 0.8

    # Clip to [0, 1]
    vis = np.clip(vis, 0.0, 1.0)
    return vis


# ----------------------------------------------------------------------
# Plot helpers
# ----------------------------------------------------------------------

def show_rgb(rgb: np.ndarray, title: str = "Demosaiced RGB"):
    """
    Plot an RGB image (float in 0..1 or any scale).
    """
    rgb_disp = rgb
    # If not already in [0,1], normalize
    if rgb_disp.dtype != np.float32 and rgb_disp.dtype != np.float64:
        rgb_disp = rgb_disp.astype(np.float32)

    if rgb_disp.max() > 1.0 or rgb_disp.min() < 0.0:
        rgb_disp = normalize_for_display(rgb_disp)

    plt.figure(figsize=(6, 6))
    plt.imshow(np.clip(rgb_disp, 0.0, 1.0))
    plt.axis("off")
    plt.title(title)


def show_rgbir_mosaic(vis: np.ndarray, title: str = "RGB-IR Mosaic Visualization"):
    """
    Plot the RGB-IR mosaic visualization image.
    """
    plt.figure(figsize=(6, 6))
    plt.imshow(np.clip(vis, 0.0, 1.0))
    plt.axis("off")
    plt.title(title)


```

```python
# ----------------------------------------------------------------------
# Main entry point (for running as a script)
# ----------------------------------------------------------------------

def main():
    # 1. Load raw frame
    raw = load_raw_frame(RAW_FILENAME, WIDTH, HEIGHT, dtype=np.uint16)

    # 2. Flip horizontally
    raw_flipped = flip_horizontal(raw)

    # 3. Demosaic to RGB
    rgb = demosaic_rgb_from_rgbir(raw_flipped)

    # 4. Build RGB-IR mosaic visualization
    rgbir_vis = build_rgbir_mosaic_vis(raw_flipped)

    # 5. Display
    show_rgb(rgb, title="Mira220 Demosaiced RGB (horizontally flipped)")
    show_rgbir_mosaic(rgbir_vis, title="Mira220 RGB-IR Mosaic (horizontally flipped)")

    plt.show()


if __name__ == "__main__":
    main()
```
